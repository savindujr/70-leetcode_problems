```python
def missingNumber(nums):
    nums.sort()

    for i,v in enumerate(nums):
        if (i!= v):
            return v-1
        if v == len(nums)-1:
            return v+1


nums = [3,0,1,2,4,5,6,7,8,9,10,11,13]
print(missingNumber(nums))
```

```python
def missingNumber(nums):
    n = len(nums)
    # Check each number from 0 to n
    for i in range(n + 1):
        if i not in nums:  # Linear search in nums for each i
            return i

# Test
nums = [3,0,1]
print(missingNumber(nums))  
```

```python
def missingNumber(nums):
    nums.sort()
    n = len(nums)
    
    # Check if 0 is missing at the beginning
    if nums[0] != 0:
        return 0
    
    # Check for gap between consecutive numbers
    for i in range(1, n):
        if nums[i] != nums[i-1] + 1:
            return nums[i-1] + 1
    
    # If no gap found, missing number is n
    return n

# Test
nums = [3,0,1]
print(missingNumber(nums)) 
```
