```python
#2 Method

def containsDuplicate(nums):
    seen = {}
    for num in nums:
        if num in seen:
            return True
        seen[num] = True
    return False

# Test
nums = [1,2,3,4,5,6,1]
print(containsDuplicate(nums))  # True
```

```python
# 3 Method Most Fundamental Brute Force Method

def containsDuplicate(nums):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                return True
    return False

# Test
nums = [1,2,3,4,5,6,1]
print(containsDuplicate(nums))
```
