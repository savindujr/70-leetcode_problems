```python
def findDisappearedNumbers(nums):
    n = len(nums)
    result = []

    #Checking each number from 1 to n
    for i in range(1,n+1):
        found = False
        # Linear Search Through Array
        for num in nums:
            if num == i:
                found = True
                break
        if not found:
            result.append(i)

    return result

nums = [4,3,2,7,8,2,3,1]
print(findDisappearedNumbers(nums))  
```

```python
# Using set function 

def findDisappearedNumbers(nums):
    n = len(nums)
    num_set = set(nums)
    result = []
    
    for i in range(1, n + 1):
        if i not in num_set:
            result.append(i)
    
    return result

# Test
nums = [4,3,2,7,8,2,3,1]
print(findDisappearedNumbers(nums))  # [5,6]
```

```python
# Boolean array extra  sapce

def findDisappearedNumbers(nums):
    n = len(nums)
    present = [False] * (n + 1)  # index 0 to n
    
    # Mark present numbers
    for num in nums:
        present[num] = True
    
    # Collect missing numbers
    result = []
    for i in range(1, n + 1):
        if not present[i]:
            result.append(i)
    
    return result

# Test
nums = [4,3,2,7,8,2,3,1]
print(findDisappearedNumbers(nums))  # [5,6]
```
